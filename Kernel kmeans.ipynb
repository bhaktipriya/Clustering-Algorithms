```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_circles
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
```

```python
#circles
X,Y=make_circles(n_samples=500,factor=.3,noise=.04)
kpca=KernelPCA(kernel="rbf", fit_inverse_transform=True, gamma=10)
X_kpca=kpca.fit_transform(X)
kmeans=KMeans(n_clusters=2, random_state=0).fit(X_kpca)
plt.title("Clustered")
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.show()
```

```python
noisy_moons=make_moons(n_samples=500,noise=.04)
X,Y=noisy_moons
kpca=KernelPCA(kernel="rbf", fit_inverse_transform=True, gamma=2)
X_kpca=kpca.fit_transform(X)
kmeans=KMeans(n_clusters=2, random_state=0).fit(X_kpca)
plt.title("Clustered")
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.show()
```
